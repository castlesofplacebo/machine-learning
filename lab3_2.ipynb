Нужно использовать регуляризацию **L1, L2 или Elastic Net**.

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

scaler = StandardScaler()

X_train = scaler.fit_transform(training_points)
```

```python
def mini_batch_l2(X, y, learning_rate=0.01, n_epochs=1000, k=40, l2=1):
    
    w = np.zeros(X.shape[1])
    b = 0
    
    for i in range(n_epochs):
        
        index = X.sample(k).index.to_list()

        X_tr = X.loc[index].values
        y_tr = y.loc[index].values
        
        for i in range(k):
              
            Lw = (-2 * X_tr[i]) * (y_tr[i] - np.dot(X_tr[i],w.T) - b) + 2 / k * l2 * w
            Lb = (-2 / k) * np.sum(y_tr[i] - np.dot(X_tr[i],w.T) - b)
            
            w = w - learning_rate * Lw
            b = b - learning_rate * Lb
        
        learning_rate = learning_rate / 1.02
        
    return w, b
```

```python
def predict(X, w, b):
    y_pred=[]
    for i in range(len(X)):
        X_test = X.values
        y = (np.dot(w, X_test[i]) + b).item()
        y_pred.append(y)
    return np.array(y_pred)
```
